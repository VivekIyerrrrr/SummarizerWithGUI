```python
import tkinter as tk
from tkinter import messagebox  # Importing messagebox for displaying messages
import nltk
from textblob import TextBlob
from newspaper import Article

# Downloading the necessary NLTK data
nltk.download('punkt')

# Global variable to track the current line to be printed for jokes
current_line = 1

def summarize():
    url = utext.get('1.0', "end").strip()  # Get the URL input from the Text widget

    try:
        article = Article(url)
        article.download()
        article.parse()
        article.nlp()

        # Update Text widgets with article information
        title.config(state='normal')
        title.delete('1.0', 'end')
        title.insert('1.0', article.title)
        title.config(state='disabled')

        author.config(state='normal')
        author.delete('1.0', 'end')
        author.insert('1.0', ', '.join(article.authors))  # Join multiple authors with comma
        author.config(state='disabled')

        summary.config(state='normal')
        summary.delete('1.0', 'end')
        summary.insert('1.0', article.summary)
        summary.config(state='disabled')

        publication.config(state='normal')
        publication.delete('1.0', 'end')
        publication.insert('1.0', str(article.publish_date))  # Convert publish_date to string
        publication.config(state='disabled')

        # Analyze sentiment using TextBlob
        analysis = TextBlob(article.text)
        sentiment.config(state='normal')
        sentiment.delete('1.0', 'end')
        sentiment.insert('1.0', f'Sentiment: {"positive" if analysis.polarity > 0 else "negative" if analysis.polarity < 0 else "neutral"}')
        sentiment.config(state='disabled')

    except Exception as e:
        messagebox.showerror("Error", f"Error while summarizing article: {str(e)}")

def Im_feeling_lucky():
    global current_line
    
    if current_line == 1:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "Why was the mobile phone wearing glasses?... Because it lost its contacts.")
        joke_text.config(state='disabled')
        current_line = 2
    elif current_line == 2:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "What did the computer have during his break time?... He had a byte!")
        joke_text.config(state='disabled')
        current_line = 3
    elif current_line == 3:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "What kind of computer sings the best?... A Dell.")
        joke_text.config(state='disabled')
        current_line = 4
    elif current_line == 4:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "I took a Microsoft Office class in high school… I Excelled in it.")
        joke_text.config(state='disabled')
        current_line = 5
    elif current_line == 5:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "What kind of books grow on trees?... Apple MacBooks.")
        joke_text.config(state='disabled')
        current_line = 6
    elif current_line == 6:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "Why do vampires use Linux?... Because they don't like Windows in their house.")
        joke_text.config(state='disabled')
        current_line = 7
    elif current_line == 7:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "My Xbox, PS4, and Switch all broke on the same day… I was inconsolable.")
        joke_text.config(state='disabled')
        current_line = 8
    elif current_line == 8:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "What game is in Schrödinger's Xbox?... Dead or Alive.")
        joke_text.config(state='disabled')
        current_line = 9
    elif current_line == 9:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "What was the only Nintendo series to really fall flat?... Paper Mario.")
        joke_text.config(state='disabled')
        current_line = 10
    elif current_line == 10:
        joke_text.config(state='normal')
        joke_text.delete('1.0', 'end')
        joke_text.insert('1.0', "How many Nintendo fans does it take to change a lightbulb?... Who knows? They’re all too busy playing with the Switch.")
        joke_text.config(state='disabled')
        current_line = 1

# Create the main application window
root = tk.Tk()
root.title("News Summarizer and Jokes")
root.geometry('800x600')  # Set window size

# Create a label and Text widget for entering URL
ulabel = tk.Label(root, text="Enter URL:")
ulabel.pack()

utext = tk.Text(root, height=1, width=100)
utext.pack()

# Button to trigger summarization
summarize_btn = tk.Button(root, text="Summarize", command=summarize)
summarize_btn.pack(pady=10)

# Button to display jokes
joke_btn = tk.Button(root, text="DON'T PRESS THE RED BUTTON", command=Im_feeling_lucky,bg="red")
joke_btn.pack(pady=5)

# Labels and Text widgets for displaying article information
tlabel = tk.Label(root, text="Title:")
tlabel.pack()

title = tk.Text(root, height=1, width=100)
title.config(state='disabled', bg='#dddddd')
title.pack()

alabel = tk.Label(root, text="Authors:")
alabel.pack()

author = tk.Text(root, height=2, width=100)
author.config(state='disabled', bg='#dddddd')
author.pack()

slabel = tk.Label(root, text="Summary:")
slabel.pack()

summary = tk.Text(root, height=10, width=100)
summary.config(state='disabled', bg='#dddddd')
summary.pack()

plabel = tk.Label(root, text="Publication Date:")
plabel.pack()

publication = tk.Text(root, height=1, width=100)
publication.config(state='disabled', bg='#dddddd')
publication.pack()

selabel = tk.Label(root, text="Sentiment:")
selabel.pack()

sentiment = tk.Text(root, height=1, width=100)
sentiment.config(state='disabled', bg='#dddddd')
sentiment.pack()

# Label and Text widget for displaying jokes
jlabel = tk.Label(root, text="DON'T PRESS THE RED BUTTON")
jlabel.pack()

joke_text = tk.Text(root, height=2, width=100)
joke_text.config(state='disabled', bg='#dddddd')
joke_text.pack()

# Start the tkinter main loop to display the window
root.mainloop()
```
